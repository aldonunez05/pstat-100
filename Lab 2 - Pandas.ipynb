## Pandas

A **data frame** is a tabular data structure with the following properties:

- Each column contains only one data type.
- Columns are indexed (typically string labels).
- Rows are indexed (typically ordinal numbers).

An index is represented by a series object, which is a 1D labelled array.

### Creating Data Frames

The `pandas` documentation provides two primary syntaxes for constructing a data frame from scratch:

1. From a **dictionary**.
2. From row-wise tuples.

#### Data Frames from Dictionaries

We can create a data frame by first creating a dictionary (a concatenation of named lists) using braces.  Here the key values will become the column names and we note that we must ensure that all of the dictionary lists are all of the same length:

```python
# Create a dictionary of athlete information
athlete_dict = {
    'name': ['John', 'Jane', 'Jim', 'Jill'],
    'age': [20, 21, 22, 23],
    'height': [180, 170, 185, 175],
    'weight': [70, 65, 80, 75],
    'team': ['A', 'B', 'A', 'B']
}
```

We can then use the `DataFrame` function to convert this dictionary into a data frame object:

```python
athlete_df = pd.DataFrame(athlete_dict)
athlete_df
```

#### Data Frames from Tuples

We can create the same data frame from a list of tuples (finite, ordered, and immutable sequence of elements).  Tuples are created using parentheses and here each tuple corresponds to a row of the data frame:

```python
athlete_tuples = [
    ('John', 20, 180, 70, 'A'),
    ('Jane', 21, 170, 65, 'B'),
    ('Jim', 22, 185, 80, 'A'),
    ('Jill', 23, 175, 75, 'B')
]
```

We can again use the `data frame` function to convert this list of tuples into a data frame object, this time including the argument `columns` whose input is a list specifying the **column names** of the data frame:

```python
athlete_df = pd.DataFrame(
    athlete_tuples, 
    columns=['name', 'age', 'height', 'weight', 'team']
    )
athlete_df
```

### Attributes

Data frames have the following built-in attributes:

1. `.shape` - dimensions (number of rows and columns).
2. `.dtypes` - column data types (float, integer, object, etc.).
3. `.size` - first dimension (number of rows).
4. `.values` - an array of all data frame entries.
5. `.columns` - column index (column names).
6. `.index` - row index (row names).

We can obtain these attributes by appending the attribute name to the data frame name using the `.` syntax.  For example, here are the attributes for our `athlete_df`:

```python
print(f"Shape: {athlete_df.shape}")
print(f"Data types: \n{athlete_df.dtypes}")
print(f"Size: {athlete_df.size}")
print(f"Values: \n{athlete_df.values}")
```

Recall these are f-strings, used to return strings whilst automatically converting data outputs.

### Indexing and Slicing

Index values are a way of referencing and retrieving specific values in multi-dimensional data structures.  Since data frames are always 2 dimensional, elements indexes have 2 values, for row and column, which are stored at the `.index` and `.columns` attributes respectively.  By default in data frames, rows are indexed by integers and columns by string labels:

```python
print(f"Column index values: \n{athlete_df.columns}")
print(f"Row index values: \n{athlete_df.index}")
```

The elements of the data frame can be retrieved using **location** `.loc[ROW-INDEX, COL-INDEX]`.  Note the difference between `python` and `R` syntax.  If we would prefer to index elements numerically (i.e. row and column numbers) we can instead use `.iloc[ROW-POS]`:

```python
print(athlete_df.loc[0, 'name'])
print(athlete_df.iloc[0, 0])
```

We can also use the `.index` and `.columns` attributes to change the values of the index, e.g. to set row names:

```python
athlete_df.index = ['Athlete 1', 'Athlete 2', 'Athlete 3', 'Athlete 4']
athlete_df
```

We can return entire columns / rows by **slicing**, that is using `:` to specify that we would like to have all of the values for that index.  We can also use `:` to specify certain ranges of values such as `a:b`:

```python
athlete_df.loc[['Athlete 1', 'Athlete 2'], :]
athlete_df.iloc[1:3, :]
```

We can take this futher and specify whole lists of index values that we would like to return, giving us complete control over that we index:

```python
athlete_df.iloc[[1,3],[0,1,2,4]]
```

### Filtering

As with `numpy` we can **filter** data frames, that is, we can index values based on logical conditions:

```python
athlete_df[athlete_df.team == "B"]
```

A less efficient version uses `.loc` but this time we specify location using a boolean vector:

```python
athlete_df.loc[athlete_df['age'] > 21, :]
```

Below is a table of some of the most common symbols used to construct logical statements in python:

| Symbol | Usage | Meaning |
|:------:|-------|---------|
| `==` | `a == b` | Does a equal b? |
| `<=` | `a <= b` | Is a less than or equal to b? |
| `>=` | `a >= b` | Is a greater than or equal to b? |
| `<` | `a < b` | Is a less than b? |
| `>` | `a > b` | Is a greater than b? |
| `~` | `~p` | Returns negation of p |
| <code>&#124;</code> | `p | q` | p OR q |
| `&` | `p & q` | p AND q |
| `^` | `p ^ q` | p XOR q (exclusive or) |

```python
athlete_df[(athlete_df.age > 21) & (athlete_df.team == "B")]
```

```python
athlete_df.drop(columns=['country'], index=['Athlete 3', 'Athlete 4'])
```


### Grouping and Aggregation

Grouping and aggregation are useful in generating data summaries, which are often important starting points in exploring a dataset.

**Aggregation** literally means 'putting together' (etymologically the word means 'joining the herd') -- in statistics and data science, this refers to data summaries like an average, a minimum, or a measure of spread such as the sample variance or mean absolute deviation (data herding!). From a technical point of view, operations that take multiple values as inputs and return a single output are considered summaries -- in other words, statistics. Some of the most common aggregations are:

* sum
* product
* count
* number of distinct values
* mean
* median
* variance
* standard deviation
* minimum/maximum
* quantiles

Pandas has built-in data frame operations that compute most of these summaries across either axis (column-wise or row-wise):

* `.sum()`
* `.prod()`
* `.mean()`
* `.median()`
* `.var()`
* `.std()`
* `.nunique()`
* `.min()` and `.max()`
* `.quantile()`

To illustrate these operations, let's consider a larger data set.  We use `head()` to view the first few rows:

```python
baby_names = pd.read_csv('data/baby_names.csv')
print(f"Shape: {baby_names.shape}")
baby_names.head()
```

We see that this is a very large data set with records of counts of names per year for each state.  To demonstrate some of these aggregations first isolate data from 1995.

```python
baby_names_1995 = baby_names[baby_names['Year'] == 1995]
baby_names_1995.head()
```

How many individuals were counted in total in 1995? We can address that by computing a sum of the counts:

```python
# n for 1995
print(baby_names_1995.Count.sum())
```

What is the typical frequency of all names in 1995? We can address that by computing the average count: 

```python
# average count for a name in 1995
print(baby_names_1995.Count.mean())
```

Suppose we would like to find the most common male baby name in 1995.  If we try to apply the maximum we run into a problem.  The output returns the maximum (alphabetically in some cases) of all columns!

```python
baby_names_1995.max()
```

We overcome this problem using **grouping**.  In general, any variable in a data frame can be used to define a grouping structure on the rows (or, less common, columns). After grouping, any data frame operations will be executed within each group, but not across groups. This can be used to generate grouped summaries, such as the maximum count for boys and girls; as a point of terminology, we'd describe this summary as 'maximum count by sex' (SUMMARY by GROUPING VARIABLE).

The `.groupby()` function defines such a structure; here is the [documentation](https://pandas.pydata.org/docs/reference/api/pandas.data frame.groupby.html). The cell below groups the `names_95` data frame by sex. Notice that when the grouped data frame is previewed with `.head()`, the first few rows are returned *for each group*.

```python
baby_names_1995_bysex = baby_names_1995.groupby('Sex')
baby_names_1995_bysex.head()
```

Now any aggregation operations applied to the grouped data frame will be applied separately to the rows where `Sex == M` and the rows where `Sex == F`. For example, computing `.sum()` on the grouped data frame will show the total number of individuals in the data for 1995 *by sex*:

```python
# number of individuals by sex
baby_names_1995_bysex.Count.sum()
```

The most frequent boy and girl names can be found using `.idxmax()` groupwise to obtain the index of the first occurrence of the maximum count for each sex, and then slicing with `.loc`:

```python
# first most common names by sex
baby_names_1995.loc[baby_names_1995_bysex.Count.idxmax(), :]
```

This is an example of chaining operations (i.e. we have first applied `Count` and then `idxmax`).  This is similar to piping in `R` and allows us to apply multiple functions one after the other on the same line of code.  In reality, we did not even need to make a new data set, we can also perform the filtering in line:

```python
baby_names[baby_names.Year == 1995].groupby('Sex').Count.sum()
```

Chains can get somewhat long, but they have the advantage of making codes more efficient, and often more readable. We did above in one step what took several lines before. Further, this chain can almost be read aloud:

"Take baby names, filter on year, *then* group by sex, *then* select name counts, *then* compute the sum."

Let's now consider computing the average counts of boy and girl names for each year 1990-1995. This can be accomplished by the following chain (notice it is possible to group by multiple variables).

```python
# average counts by sex and year
baby_names[baby_names.Year <= 1995].groupby(['Year', 'Sex'])['Count'].mean()
```

This display is not ideal. We can 'pivot' the table into a wide format by adding a few extra steps in the chain: change the indices to columns; then define a new shape by specifying which column should be the new row index, which should be the new column index, and which values should populate the table.

```python
# average counts by sex and year
baby_names[baby_names.Year <= 1995].groupby(['Year', 'Sex'])['Count'].mean().reset_index().pivot(index = 'Sex', columns = 'Year', values = 'Count')
```

**Style comment: break long chains over multiple lines with indentation.** The above chain is too long to be readable. To balance the readability of codes with the efficiency of chaining, it is good practice to break long chains over several lines, with appropriate indentations. Here is a better-styled version of the previous cell:

```python
# better style
baby_names[baby_names.Year <= 1995].groupby(
    ['Year', 'Sex']
    )['Count'].mean(
    ).reset_index(
    ).pivot(
        index = 'Sex', 
        columns = 'Year', 
        values = 'Count'
    )
```

Here are some rules of thumb on style.

* Separate comparisons by spaces (`a<b` as `a < b`)
* Split chains longer than 30-40 characters over multiple lines
* Split lines between delimiters `(`, `)`
* Increase indentation for lines between delimiters
* For chained operations, try to get each step in the chain shown on a separate line
* For functions with multiple arguments, split lines so that each argument is on its own line

### Merging

In real data work, information about the same “thing” is often split across multiple tables. Pandas supports SQL-style joins using `pd.merge`.

**Key ideas (read before starting Problem 3):**

- A **key** is a column (or set of columns) that identifies which rows “match” across tables.
- The most common syntax is:

`pd.merge(left_df, right_df, on='key_column', how='left')`

- `how=` controls which rows are kept:
    - `inner`: only matching keys in *both* tables
    - `left`: keep all rows from the left table, match what you can from the right
    - `right`: keep all rows from the right table
    - `outer`: keep all rows from both (fills missing values with `NaN`)

**Helpful tips:**

- If the key column names differ, use `left_on=...` and `right_on=...`.
- If both tables share non-key column names, use `suffixes=('_left', '_right')`.
- A good sanity check is to compare the number of rows before/after the merge and to inspect a few merged rows with `.head()`.

### Reshaping

You’ve already seen how to make a “wide” table using `pivot(...)`. Another common reshape is **wide → long**, which is usually done with `.melt(...)`.

**When to use what:**

- `pivot(index=..., columns=..., values=...)`: long → wide (creates a spreadsheet-like table)
- `reset_index()`: turns index levels back into ordinary columns
- `melt(...)`: wide → long (creates one row per combination of identifiers)

A typical workflow is:

1. Compute a summary in long format
2. Pivot it to wide format for readability
3. (Optionally) melt it back to long format for plotting/modeling
